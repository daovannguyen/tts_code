```python
def split_text(text):
  space = 500
  arr = []
  number = int(len(text)/space)
  for i in range(number + 1):
    arr.append(text[i*space:(i+1)*space])
  return arr
```

```python
def tach_cau(path):

  # Đọc file và tiền xử lí
  with open(path, 'r', encoding='utf-8') as f:
    text = f.read()
  for i in range(10):
    text = text.replace('\n\n', '\n')
    text = text.replace('  ', ' ')
  a = text.split('.')

  # Tách nhỏ file
  arr = []
  element = ''
  for i in a:
    if len(urllib.parse.quote(i)) > 2500:
      if len(element) == 0:
        adasjdsa = 0
      else:
        arr.append(element)
      arr = arr + split_text(i)
      element = ''
    elif len(urllib.parse.quote(element) + urllib.parse.quote(i)) < 2500:
      element = element + i
    else:
      arr.append(element)
      element = i
  arr.append(element)
    
  return arr
```

```python
def str_to_mp3(str, path):
  str_start = 'https://texttospeech.responsivevoice.org/v1/text:synthesize?text=' 
  str_end = '&lang=vi&engine=g1&name=&pitch=0.5&rate=0.5&volume=1&key=0POmS5Y2&gender=female'
  a = urllib.parse.quote(str)
  url = str_start + a + str_end
  n_html.download_from_url(url, path)
```

```python
def gop_file(save_path):
  try:
    os.remove(save_path)
  except:
    sajdjskadka = 0

  a = os.listdir('/content/1')
  a.sort()
  for i in tqdm(a):
    path = os.path.join('/content/1', i)
    with open(path, 'rb') as f:
      b = f.read()
      
    with open(save_path, 'a+b') as f:
      f.write(b)
```

```python
def text_to_mp3(file_path, save_path):
  try:
    shutil.rmtree('/content/1')
  except: 
    desadasjaks = 0
  os.mkdir('/content/1')

  arr = tach_cau(file_path)
  for i, x in tqdm(enumerate(arr)):
    path = os.path.join('/content/1', str(i).zfill(4))
    str_to_mp3(x, path)

  gop_file(save_path)
```

```python
for i in tqdm(os.listdir('/content/drive/MyDrive/tts/text')):
  try:
    o = os.path.join('/content/drive/MyDrive/tts/tts_request_from_web/text', i)
    n = os.path.join('/content/drive/MyDrive/tts/tts_request_from_web/mp3', i[:-4] + '.mp3')
    text_to_mp3(o, n)
    os.remove(o)
  except:
    print('sjakakdjad')
```
