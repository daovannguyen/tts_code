```python
def create_mp3(file):
  with open(file, 'r', encoding = 'utf-8') as myfile:
    text = myfile.read()
  output = gTTS(text=text, lang='vi', slow=False)
  output.save('/content/' + file[:-4] + '.mp3')
```
