**Di chuyển TXT bình thường**

```python

path = '/content/drive/MyDrive/tts/folder/mp3'
folder = os.listdir(path)
x = []
for i in tqdm(folder):
  txt = os.listdir(os.path.join(path, i, i))
  mp3 = []
  for j in os.listdir(os.path.join(path, i)):
    if j.find('.mp3') > -1:
      mp3.append(j)
  for j in mp3:
    txt.remove(j[:-4] + '.txt')
  x = x + txt
print(len(x))
x.sort()
print(x)
```

```python
data = []
for i in x:
  a = os.path.join('/content/drive/MyDrive/tts/folder/mp3', i[:4], i[:4], i)
  b = os.path.join('/content/tts/text', i)
  c = os.path.join('/content/drive/MyDrive/tts/text', i)
  data.append([i, a, b, c])
print(data[:3])
print(len(data))
```

```python
for i in tqdm(data):
  shutil.copy(i[1], i[2])
```
